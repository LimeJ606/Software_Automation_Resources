# The Cost Function: Step-by-Step

In the previous exercise, you manually changed the line to make it fit the data. But how does a computer know when a line is a "Good Fit"? 

It uses a scorecard called the **Cost Function**. The Cost Function measures how wrong our predictions are. 
**Our goal is to make the cost as close to zero as possible.**

Let's build this scorecard one piece at a time.

* **Loss:** The error for a **single** data point. (How far off was one guess?)
* **Cost:** The average error over the **entire** dataset. (How bad is the line overall?)

$\text{Error} = \text{Prediction} - \text{Actual Target}$

If you did the step above correctly, your error was `-200`. 

Negative numbers are annoying when trying to find an "average" error, because a `-200` error and a `+200` error might cancel each other out to zero, making the computer think it did a perfect job! 

To fix this, we **square the error** (multiply it by itself). This does two things:
1.  It makes all errors positive.
2.  It heavily penalizes really big mistakes.

$\text{Squared Error} = (f_{w,b}(x) - y)^2$

1.  **$\sum$ (Sum):** Add up the squared errors for every single data point.
