# Numpy: Tracking High School Habits
A Python library used for working with arrays. In this exercise, we will use a NumPy array, which is a powerful N-dimensional array object, to analyze weekly habits over a 12-week semester: Study Hours vs. Screen Time.

The [Numpy Documentation](https://numpy.org/doc/2.2/user/absolute_beginners.html#) is very user friendly.

```python
# importing Numpy library
import numpy as np
```

```python
# Create a list representing hours studied per week for 12 weeks
study_hours_list = [5, 4, 6, 7, 8, 5, 10, 12, 9, 6, 8, 11]

# creating numpy array
study_array = np.array(study_hours_list)

print(f"List in python :\n{study_hours_list}\n")

print(f"Numpy Array in python :\n{study_array}")
```

```python
# Create a second list representing screen time (phone/gaming) per week
screen_time_list = [18, 20, 17, 15, 12, 19, 10, 8, 14, 16, 15, 12]
screen_array = np.array(screen_time_list)

# creating multi-dimensional numpy arrays to compare them
# This stacks them as rows (2 rows, 12 columns)
habits_rows = np.array([study_hours_list, screen_time_list])

# This pairs them side-by-side as columns (12 rows, 2 columns)
habits_columns = np.column_stack((study_array, screen_array))

print(f"Screen Time List :\n{screen_time_list}\n")

print(f"2x12 Numpy Multi-dimensional Array (Rows) :\n{habits_rows}\n")
print(f"12x2 Numpy Multi-dimensional Array (Columns) :\n{habits_columns}\n")
```

#### Array attributes

ndim, shape, size, and dtype are attributes of an Numpy array.

```python
# An integer that tells us how many dimensions the array have.
print(f"Number of dimensions (study_array): {study_array.ndim}")
print(f"Number of dimensions (screen_array): {screen_array.ndim}")
print(f"Number of dimensions (habits_rows): {habits_rows.ndim}")
print(f"Number of dimensions (habits_columns): {habits_columns.ndim}")
```

```python
# A tuple of non-negative integers that specify the number of
# elements along each dimension called the shape.
print(f"Shape of study_array: {study_array.shape}")
print(f"Shape of screen_array: {screen_array.shape}")
print(f"Shape of habits_rows: {habits_rows.shape}")
print(f"Shape of habits_columns: {habits_columns.shape}")
```

```python
# An object describing the type of the elements in the array.
print(f"Data type of study_array: {study_array.dtype}")
print(f"Data type of screen_array: {screen_array.dtype}")
print(f"Data type of habits_rows: {habits_rows.dtype}")
print(f"Data type of habits_columns: {habits_columns.dtype}")
```

```python
# The fixed, total number of elements in array
print(f"Size of study_array: {study_array.size}")
print(f"Size of screen_array: {screen_array.size}")
print(f"Size of habits_rows: {habits_rows.size}")
print(f"Size of habits_columns: {habits_columns.size}")    
```

#### Reshaping an Array

Reshaping will give a new shape to an array without changing the data. For example, instead of looking at 12 individual weeks, we can reshape our 12 weeks of study data into 3 months (where each month has 4 weeks). 3 rows x 4 columns = 12 elements.

```python
# Reshape the 12-week array into 3 months (rows) of 4 weeks (columns)
print(f"Original 12-week Study Array:\n{study_array}\n")
monthly_study_reshape = study_array.reshape(3,4) # 3*4 = size of array
print(f"Reshaped array (3 months, 4 weeks each):\n{monthly_study_reshape}")
```

```python
# Reshape the array to 2D array from a 1D array with 1 row and 12 columns
reshape_1x12 = study_array.reshape(1,12)
print(f"Reshaped array (1x12):\n{reshape_1x12}")
```

```python
# Flatten our 3-month (2D) array back into a continuous 12-week (1D) array
print(f"Current 2D array:\n{monthly_study_reshape}\n")
flatten = monthly_study_reshape.flatten()
print(f"Flattened array:\n{flatten}")
```
