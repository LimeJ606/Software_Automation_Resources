### Exercise 1: Setting up the DataFrame
1. Import the `numpy` and `pandas` libraries.
2. Create two lists: `student_ids` containing the numbers 101 through 110, and `test_scores` containing ten random scores.
3. Create a pandas DataFrame using these two lists. Add a third column called "Entry Date" that records the current date and time.
4. Print the entire DataFrame.

```python
# Write your code for Exercise 1 here
import numpy
import pandas
student_ids = [101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 111, 112]
test_scores = [5, 64, 83, 93, 98, 34, 37, 84, 56, 21, 43, 34]

dataframe = pandas.DataFrame(
    {
        "student id": student_ids,
        "test score": test_scores, 
        "at time": pandas.Timestamp.now(),
        })

print(dataframe)
dtypes = dataframe.dtypes
print(dtypes)
```

```python
# Write your code for Exercise 3 here
print(dataframe.describe())
shape = dataframe.shape
print(shape)
dimensions = dataframe.ndim
print(dimensions)
size = dataframe.size
print(size)
row = len(dataframe)
print(row)
column = len(dataframe.columns)
print(column)
```

## Solutions
Below are the solution code cells for the exercises. Try to write your own code before peeking!

```python
# SOLUTION 1
import numpy as np
import pandas as pd

student_ids = [101, 102, 103, 104, 105, 106, 107, 108, 109, 110]
test_scores = [85, 92, 78, 90, 88, 76, 95, 89, 82, 91]

df_students = pd.DataFrame(
    {
        "Entry Date": pd.Timestamp.now(),
        "Student ID": student_ids,
        "Test Score": test_scores
    }
)

print("Full DataFrame:\n", df_students)
```

```python
# SOLUTION 2
print("\nData Types:\n", df_students.dtypes)
print("\nFirst 5 rows:\n", df_students.head())
print("\nLast 5 rows:\n", df_students.tail())
```

```python
# SOLUTION 3
print("\nSummary Statistics:\n", df_students.describe())
print("\nShape:", df_students.shape)
print("Dimensions:", df_students.ndim)
print("Total Elements:", df_students.size)
print("Number of Rows:", len(df_students))
print("Number of Columns:", len(df_students.columns))
```
