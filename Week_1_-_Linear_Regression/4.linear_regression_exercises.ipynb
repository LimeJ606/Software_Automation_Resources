### The Magic Formula

At the heart of Linear Regression is one very famous, but very friendly, equation:

$y = wx + b$

Think of it like a recipe. You plug your data ($x$) into the machine, tweak two dials ($w$ and $b$), and the machine spits out a prediction ($y$).

| Symbol | Name | What it does in plain English |
| :--- | :--- | :--- |
| **y** | Target | The prediction we are trying to make (e.g., the price of a house). |
| **x** | Feature | The data we already know (e.g., the size of the house). |
| **w** | Weight (Slope) | The **Seesaw**: it determines how steep the line is. |
| **b** | Bias (Intercept) | The **Elevator**: it moves the entire line straight up or down. |

### Exercise 3: Translating Math to Code

Now that you know how the formula behaves visually, let's write it in Python. 
We need to tell the computer how to calculate the prediction (`y_pred`) for every data point (`x`) we give it.

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    """
    X = x.shape[0]
    y_pred = np.zeros(X)
    
    # This loop goes through every single data point we have
    for i in range(X):
        # TODO: Implement the magic formula: y = wx + b
        # Replace '0' with the formula. 
        # Hint: In Python, multiply is '*' and add is '+'. 
        # The current data point is represented by x[i]
        
        y_pred[i] = 0 
        
    return y_pred
```

### A Quick Lesson on "Fit"

In the real world, data is messy. You will almost never be able to make a line touch every single dot perfectly. 

* **Underfitting:** The line doesn't learn the pattern at all.
* **Overfitting:** Trying too hard to touch every dot, resulting in a crazy shape that won't work on new data.
* **Good fit (Intuition):** A clean line that slices right through the "middle" of the data trend.

```python
# Run this cell to visualize Good Fit vs Bad Fit
m = np.sort(generate_data(-10, 10, -10, 10, 8, 0.3))

fig, ax = plt.subplots(1,3, figsize=(15, 5))
ax[0].scatter(m[0], m[1], color='blue', s=100)
ax[0].plot(m[0], m[1], color='red')

x = np.linspace(10, -10, 100)
y = -x**2 + 4*x + 2  
ax[2].plot(x, y, color='red')
x = x + np.random.uniform(-5, 5, size=x.shape)
y = y + np.random.uniform(-10, 10, size=y.shape)
ax[1].scatter(x, y, c='b')
ax[2].scatter(x, y, c='b')

x_lin = np.array([-10,10])
tmp_f_wb = compute_model_output(x_lin, 10, -10)
ax[1].plot(x_lin, tmp_f_wb, c='r')

ax[0].title.set_text("Overfitting")
ax[1].title.set_text("Underfitting")
ax[2].title.set_text("Good Fit / Intuition")
for a in ax.flat:
    a.set_xticks([])
    a.set_yticks([])

plt.show()
```
