### Step 2: The Magic Formula

To make a prediction, computers use one simple recipe:

$y = wx + b$

Think of it like a machine with two dials (`w` and `b`). You put your data (`x`) into the machine, turn the dials, and the machine spits out a prediction (`y`).

* **y (Target):** The prediction we want (like the price of a house).
* **x (Feature):** The data we know (like the size of the house).
* **b (Bias):** We call this the **Elevator**. It moves the line up and down.
* **w (Weight):** We call this the **Seesaw**. It tilts the line up and down.

### Step 5: Translating Math to Python

When a computer looks at a list of data, it looks at one item at a time. 
In Python code, `x[i]` just means "the current piece of data I am looking at".

````python
def compute_model_output(x, w, b):
    X_total_items = x.shape[0]
    y_pred = np.zeros(X_total_items)
    
    # This loop goes through every single data point we have, one by one
    for i in range(X_total_items):
        
        # TODO: Delete the '0' below and type out the formula.
        # You need to multiply w by x[i], and then add b.
        y_pred[i] = 0 
        
    return y_pred

print("Function saved!")
````

### Step 7: Welcome to the Real World

In real life, data is messy. You will almost never be able to make a line touch every single dot perfectly. 

* **Underfitting:** The line is too simple and misses the trend entirely.
* **Overfitting:** The line wiggles like crazy to touch every dot. It works here, but will fail on new data.
* **Good fit:** A clean, straight line that slices right through the "middle" of the data cloud.

````python
# Run this cell to look at the differences visually
m = np.sort(generate_data(-10, 10, -10, 10, 8, 0.3))

fig, ax = plt.subplots(1,3, figsize=(15, 5))
ax[0].scatter(m[0], m[1], color='blue', s=100)
ax[0].plot(m[0], m[1], color='red')

x = np.linspace(10, -10, 100)
y = -x**2 + 4*x + 2  
ax[2].plot(x, y, color='red')
x = x + np.random.uniform(-5, 5, size=x.shape)
y = y + np.random.uniform(-10, 10, size=y.shape)
ax[1].scatter(x, y, c='b')
ax[2].scatter(x, y, c='b')

x_lin = np.array([-10,10])
tmp_f_wb = compute_model_output(x_lin, 10, -10)
ax[1].plot(x_lin, tmp_f_wb, c='r')

ax[0].title.set_text("Overfitting (Too crazy)")
ax[1].title.set_text("Underfitting (Too simple)")
ax[2].title.set_text("Good Fit (Just right)")
for a in ax.flat:
    a.set_xticks([])
    a.set_yticks([])

plt.show()
````
